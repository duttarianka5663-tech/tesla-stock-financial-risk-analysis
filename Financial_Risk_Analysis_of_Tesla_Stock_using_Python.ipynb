```python
import pandas as pd    #Load The Dtaset

data = pd.read_csv("tsla_sample_data.csv")
data.head()
```

```python
data.info()
data.describe()        #Basic Data Understanding
```

```python
data['Date'] = pd.to_datetime(data['Date'])        #Convert Date Column
data.set_index('Date', inplace=True)
```

```python
data = data[['Close']]                     #Focus on Closing Price
```

```python
data['Daily Return'] = data['Close'].pct_change()  #Calculate Daily Returns
data.head()
```

```python
volatility = data['Daily Return'].std()     #Calculate Risk (Volatility)
print("Volatility (Risk):", volatility)
```

```python
import matplotlib.pyplot as plt           #Plot Stock Price

data['Close'].plot(figsize=(10,5))
plt.title("Tesla Stock Price Trend")
plt.show()
```

```python
data['Daily Return'].plot(figsize=(10,5))   #Plot Daily Returns
plt.title("Daily Returns Over Time")
plt.show()
```

```python
data['Daily Return'].hist(bins=50)
plt.title("Return Distribution")            #Histogram (Risk Distribution)
plt.show()
```

```python
data['Rolling Volatility'] = data['Daily Return'].rolling(window=30).std()

data['Rolling Volatility'].plot(figsize=(10,5))          #Shows how risk changes over time
plt.title("30-Day Rolling Volatility")
plt.show()
```
